```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Abhi9\Downloads\linear_regression_train.csv")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Feature_1",y="Target",data=data)
plt.xlabel('values')
plt.ylabel('frequency')
plt.grid(axis='both',linestyle='--',alpha=0.7)
plt.title('plot')
plt.show()
```
