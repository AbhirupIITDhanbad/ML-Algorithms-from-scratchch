```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Generate random data
np.random.seed(42)
data = np.random.normal(50, 15, 200)  # Mean=50, Std Dev=15, Sample=200
data
```

```python
plt.figure(figsize=(10,5))
plt.hist(data,bins=10,color='yellow',edgecolor='red')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# for some critical dataset use seaborn library
import seaborn as sns
plt.figure(figsize=(10,5))
sns.histplot(data,bins=10,kde=True,color='red')
plt.xlabel('values')
plt.ylabel('frequency')
plt.grid(axis='both',linestyle='--',alpha=0.7)
plt.show()
```
